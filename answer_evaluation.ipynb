```python
import pandas as pd
import string
import nltk
from nltk import word_tokenize, sent_tokenize
```

```python
import spacy
nlp = spacy.load("en_core_web_lg")
```

```python
def get_sim_score(text1, text2):
    tokens1 = nlp(text1)
    tokens2 = nlp(text2)
    sim_score = tokens1.similarity(tokens2)
    return sim_score
```

```python
text1 = "A barcode is a series of light and dark bars of differing widths."
text2 = "A barcode consists of light and dark bars of different widths, with a code number printed below the bars."
text3 = "With a code number printed underneath."
text4 = "A barcode is a series of light and dark bars of differing widths with a code number printed underneath made up of \
         country of origin code, manufacturer code, the product code, a check digit."
```

```python
get_sim_score(text4, text3)
```

```python
st_ans1 = "Each product has a barcode on a label which can be scanned at checkout and will tell the EPOS system what is being bought."
gl_1 = "A barcode is a series of light and dark bars of differing widths"
gl_2 = "With a code number printed underneath"
gl_3 = "Made up of country of origin code, manufacturer code, the product code, a check digit."
```

```python
def jaccard_similarity(text1, text2):
    s1 = set(text1.split())
    s2 = set(text2.split())
    return len(s1.intersection(s2)) / len(s1.union(s2))

text1 = "A dog sat on the rat ok fine"
text2 = "A cat sat on the mat"
jaccard_similarity(text1, text2)
```
